## analysis(charts)

### gender distribution female, male

```python
plt.figure(figsize=(3,5)) #bar ka height width
# to create level i.e valaue of the bar

ax= sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

# 2. parent education impact on courses

```python
gb= df.groupby("ParentEduc").agg({"MathScore":'mean' ,"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

##### whenever we get different datatype column like here
 1. Numeric 
           -mathscore 
            -readingscore
            -writing score
 2.String 
           -parenteducation


## use HEATMAP for analysing

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot=True)
plt.show()
# to get lebeled inside the boxes - annot-True
```

###### light color -low marks
# ex:  jinke parents high school kiye h unke bacche low marks avg getting
#     jo parents master degree kiya h unka reading writing me highest h
